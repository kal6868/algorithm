```python
def solution():
    H, M = map(int, input().split())
    time = (H*60) + M
    alarm = time - 45

    if alarm > 0:
        h = alarm // 60
        m = alarm % 60
    else:
        alarm = (24*60) + alarm
        h = alarm // 60
        m = alarm % 60
    if h == 24:
        h = 0
    return h, m
```

```python
for _ in range(3):
    print(solution())
```
