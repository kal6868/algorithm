# 문제

* 정수 n이 매개변수로 주어집니다. 다음 그림과 같이 밑변의 길이와 높이가 n인 삼각형에서 맨 위 꼭짓점부터 반시계 방향으로 달팽이 채우기를 진행한 후, 첫 행부터 마지막 행까지 모두 순서대로 합친 새로운 배열을 return 하도록 solution 함수를 완성해주세요.

# 제한 조건

* n은 1 이상 1,000 이하입니다.

```python
def solution(n):
    space = [[0]*n for _ in range(n)]
    num = 1
    x, y = -1, 0
    answer = []

    for i in range(n):
        for _ in range(i, n):
            if i % 3 == 0:
                x += 1
            elif i % 3 == 1:
                y += 1
            elif i % 3 == 2:
                x -= 1
                y -= 1
            space[x][y] = num
            num += 1
    for rows in space:
        for ele in rows:
            if ele != 0:
                answer.append(ele)
    return answer
```

```python
print(solution(4))
print(solution(5))
print(solution(6))
```

# 풀이 과정 및 주의사항

* n개의 0으로 채워진 n개의 리스트를 먼저 만들고 각 위치에 해당하는 값을 넣는 과정을 거친다
* 이중 for문을 사용하여 이동하는 횟수를 1회씩 줄여나갈 수 있다
* 움직이는 방향이 3가지이므로 3으로 나눈 몫을 사용하여 x,y 값을 조정하여 이동 방향을 조정한다
